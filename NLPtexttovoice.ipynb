```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
import nltk.corpus
```

```python
#Tokenization
from nltk.tokenize import word_tokenize
```

```python
chess = " Samy Raina is the best chess streamer in the world"
```

```python
word_tokenize(chess)
```

```python
#sentence tokenizer
from nltk.tokenize import sent_tokenize
```

```python
chess2 = "Samay Raina is the best chess streamer in the world. Sagar Shah is the best chess coach in the world"
```

```python
sent_tokenize(chess2)
```

```python
#Checking the number of tokens
len(word_tokenize(chess))
```

```python
#bigrams and n-grams
astronaut = "Can anybody here me or am I talking to myself? My mind is running empty in the search for somone else"
```

```python
atronaut_token = (word_tokenize(astronaut))
```

```python
list(nltk.bigrams(atronaut_token))
```

```python
list(nltk.trigrams(atronaut_token))
```

```python
list(nltk.ngrams(atronaut_token,5))
```

```python
#Stemming
from nltk.stem import PorterStemmer
```

```python
my_stem = PorterStemmer()
```

```python
my_stem.stem('eating')
```

```python
my_stem.stem('going')
```

```python
my_stem.stem('shopping')
```

```python
#pos.tagging
```

```python
tom = "Tom Hanks is the best actor in the world"
```

```python
tom_token = word_tokenize(tom)
```

```python
nltk.pos_tag(tom_token)
```

```python
#Named entity recognition
```

```python
from nltk import ne_chunk
```

```python
president = " Ramanath Kovind is the current president of India"
```

```python
president_token = word_tokenize(president)
```

```python
president_pos = nltk.pos_tag(president_token)
```

```python
print(ne_chunk(president_pos))
```
